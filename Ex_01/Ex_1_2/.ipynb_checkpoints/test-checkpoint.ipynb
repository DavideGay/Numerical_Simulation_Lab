```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as stats
```

```python
def normal(x, mu, sigma):
    return (1/(sigma*math.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)/sigma)**2)
```

```python
N=10000
M=100

test=np.zeros(N)
for i in range(N):
    t=np.random.rand(M)
    test[i]=np.sum(t)/M
```

```python
mean=0.5
variance=(1/12)/M

pos=np.linspace(mean-3*math.sqrt(variance),mean+3*math.sqrt(variance), 1000)
nor=normal(pos, mean, math.sqrt(variance))
```

```python
nbins=50

result=plt.hist(test, bins=nbins)  #"result" here is an array with elements n, bins, ...
binwidth=result[1][1]-result[1][0]   # bin width: result[1] gives access to the array "bins"
scale=N*binwidth
plt.plot(pos, nor*scale)
plt.show()
```
