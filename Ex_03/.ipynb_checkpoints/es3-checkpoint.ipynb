# Exercise 3

In this exercise I want to numerically verify the analytical solution to the Option Pricing problem provided by the Black-Scholes theory for the correct pricing of European Call and Put options.

Options are contracts where one end assumes an obligation, and the other obtains a right. **European options**, also called **plain vannilla options**, are the most basic type of option.
The seller of the option, the *writer*, grants the buyer, the *holder*, the right to purchase (**call option**) or sell (**put option**) to them an asset with a current -variable- spot price $S(t)$ for a predetermined price of $K$ called *strike price* at the *expiry date* $T$. The holder will choose whether to buy or sell depending on the profit.

For a Call option, the buyer has the right to purchase the asset at the strike price, if at the time of expiration the market price $S(t)$ of the asset exceeds the strike price, obtaining a profit of $S(t)-K$. Otherwise, the buyer does not exercise the right, and the seller of the option earns the option *premium*, i.i fee.

For a Put option, the buyer has the right to sell the asset at the strike price, if at the time of expiration the market price $S(t)$ of the asset is less than the strike price, obtaining a profit of $K-S(t)$.

### Brownian motion

Black-Scholes theory assumes that the evolution of the price of an asset over time is stochastic and that it shows behavior similar to that of a Geometric Brownian Motion (GBM) with constant risk-free interest rates $r$ (equivalent to Brownian drift) and volatility $\sigma$.

The stochastic equation that describes the price trend of an asset is the following:
$$
dS(t)=\mu S(t)dt + \sigma S(t)dW(t)
$$

This equation takes into account two main components that influence the change in price over time.

The first term, $\mu S(t)dt$, represents the drift or deterministic growth of the asset. Here, $\mu$ represents the interest rate or expected return of the asset, while $S(t)$ it represents the price of the asset at the time $t$.

The second term, $\sigma S(t)dW(t)$, describes the random fluctuations in the price of the asset. Here, $\sigma$ represents the volatility of the asset,$dW(t)$ represents the infinitesimal increase of a Brownian process. This term represents unpredictable fluctuations in the price of the asset, which are influenced by the market and random factors.

It is important to note that the equation described above is valid for a Markov process.

### Direct sampling

given an asset price $S\sim GBM(\mu,\sigma^2)$, with initial value $S(0)$, then
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    where $W(t)\sim\mathcal{N}(0,t)$.

### Discretized path sampling

Since the increments of $W$ are independent and normally distributed, the simple recursive procedure for sampling discretized values of $S\sim GBM(\mu,\sigma^2)$ at $t_1 < t_2 < â€¦ < t_n$ is:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$

### Black-Scholes analytic solution

Black-Scholes  analytic solution for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

## Simulations

By means of a Monte Carlo simulation, we compute at time $t=0$ the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$.

We will be performing the simulation via both direct sampling and by sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$ and will compare the result with the value obtained from the analytical solution.

We will be  using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

Simulations will be performed with data blocking: 100 blocks of 100 steps.

```python
filepath='es3_1/outFile1.txt'

call_put=np.loadtxt(filepath, usecols = (0, 2, 4, 6))
errors=np.loadtxt(filepath, usecols = (1, 3, 5, 7))

x=np.arange(100)
```

```python
def make_plot(ax, x, y, err, analytic, flag):
    ax.errorbar(x, y, err)
    ax.grid(True)
    ax.axhline(y=analytic, color='r', linestyle='-')
    ax.set_xlabel('N blocks')
    ax.set_ylabel(flag)
```

```python
flags=['Call price direct', 'Put price direct', 'Call price discrete', 'Put price discrete']
analytic_array=[14.975790778311286, 5.4595325819072364, 14.975790778311286, 5.4595325819072364]

fig, axs = plt.subplots(4, 1, layout='constrained', figsize=(6, 15))
for i in range(4):
    make_plot(axs[i], x, call_put[:,i], errors[:,i], analytic_array[i], flags[i] )
plt.show()
```
