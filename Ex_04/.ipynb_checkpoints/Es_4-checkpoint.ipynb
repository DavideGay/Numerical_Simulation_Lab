# Exercise 4

The goal of this exercise is to investigate the properties of a system of particles in a microcanonical ensemble by using Molecular Dynamics (MD) simulations.

### The model

The **microcanonical ensemble** is a statistical ensemble in which the total energy, the number of particles and the volume are constant (hence why is it also referred to as NVE).

In the context of this exercise we intend to simulate a NVE system in which the reciprocal interaction between each pair of particles is described by a **Lennard-Jones** potential:

$$ V_{LJ}(|\vec{r}_i -\vec{r}_j|) =  4\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
 \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] $$
 
 where $\sigma$ and $\epsilon$ are two parameters of length and energy respectively (more to follow).
 
In particular we will perform simulations of Argon: 
$\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu

### The Verlet algorithm

The Verlet algorithm is a numerical method used to integrate Newton's equations of motion for a system of particles. To calculate the new position for all particles at each time the system uses the present time configuration as well as the previous configuration:
$$ \vec{r}(t+\delta t) \cong 2\vec{r}(t) - \vec{r}(t-\delta t) + \delta t^2 \vec{a}(t)$$

Here $\vec{a}$ is the acceleration vector.

The Verlet algorithm has relevant benefits:
- it conserves total energy, even with relatively large time steps
- it is efficient
- it is time reversible

### Periodic Boundary Conditions (PBC) and cutoff radius

Since simulating a macroscopical sized system (magnitude order $10^{23}$ particles) is computationally inaccessible, **Periodic Boundary Conditions** will be used:
we will only simulate a cell of volume $V=N_{part}/\rho$; this cell and its content (all the information about particles) will be considered as exactly replicated infinitely many times in 3D space; each time a particle passes through one side of a cell it will enter the neighboring one.
However, each particle will only interact with only one image - the closest one - of each of the other $N_{part}-1$ different particles.

Furthermore, we will implement a **cutoff radius** $r_c$:
if the distance between two particles (once PBC have been applied) is $r>r_c$ the interaction will be ignored.
Clearly, the cutoff radius will be different for each phase of Argon.

### Units

Reduced units will be used:
- unit of length $\sigma$,
- unit of energy $\varepsilon$,
- unit of mass $m$

From these basic units, we can define all the other relevant ones:
- distance $r^\star = r / \sigma$,
- Lennard-Jones potential $V^\star = 4 \left[ \left(1/r^\star\right)^{12} - \left(1/r^\star\right)^{6}\right]$,
- time $t^\star = t/\sigma\sqrt{m/\varepsilon}$,
- temperature $T^\star = k_B T /\varepsilon$,
- potential energy $V^\star = V/\varepsilon$,
- pressure $P^\star = P \sigma^3/\varepsilon$,
- density $\rho^\star = \rho\sigma^3$

### Phase parameters

1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

A time step $\delta=0.0005$ will be used all throughout the following simulations.

# Equilibration

We need to reach an equilibrium configuration before performing any measurements.
In this section we will investigate how and after how many simulation steps the system will reach equilibrium.

First, we recall that in a microcanonical ensemble temperature is **not** fixed. Therefore, to reach the desired "target" temperature for each simulation, we will need to identify a starting temperature that, as the simulation progresses, will evolve and ultimately stabilise around the target temperature.

The following plots will show, for each phase, 3 runs at different starting temperatures; each run will start from a non equilibrium configuration - an fcc crystal lattice - and will use blocks of 1 step.

```python
fig, axs = plt.subplots(3, 2, figsize = (15, 15))
temps_array=[1.5, 1.55, 1.6, 1.95, 1.98, 2, 0.9, 1, 1.1]
phases=['SOLID','LIQUID','GAS']
target_t=[0.8,1.1,1.2]
cut=5000

for phase in range(3):
    if phase==2:
        cut=50000
    for t in range(3):
        x, solid_instant_t, solid_ave_t, solid_err_t =np.loadtxt('../es4/DATA/'+phases[phase]+'/equilibration/temperature_' + str(temps_array[3*phase+t]) + '.dat',usecols=(0,1,2,3), unpack=True)
        axs[phase,0].plot(x[:cut], solid_instant_t[:cut], label='$ T^*_{start}= $' + str(temps_array[3*phase+t]))
        axs[phase,1].plot(x, solid_ave_t, label='$ T^*_{start}= $' + str(temps_array[3*phase+t]))

    for i in range(2):
        axs[phase,i].axhline(target_t[phase], linestyle='dotted', color='r', label='target temperature')
        axs[phase,i].set_xlabel('Steps')
        axs[phase,i].set_ylabel('$T^*$')
        axs[phase,i].legend()
        axs[phase,i].grid(True)
    axs[phase,0].set_title(phases[phase]+' phase, instant temperature')
    axs[phase,1].set_title(phases[phase]+' phase, moving average of temperature')

plt.show()
```

For the solid and liquid phase we can clearly see that, after an initial steep decrease, the temperature of the system stabilizes around the target temperature. We decide that 3000 simulation steps will be sufficient to bring the system to the desired equilibrium temperature.

In the gas phase, the system instead takes a significantly longer time to stabilize. We set the equilibration steps to 20000.

Based on the information provided by the graphs above, for each phase the: starting temperature that best approaches the targeted temperature is thus identified as being:

1. solid phase: $ T^*_{start} = 1.55 $
2. liquid phase: $ T^*_{start} = 1.98 $
3. gas phase: $ T^*_{start} = 1.0 $

# Equilibrium properties

### Implementing Pressure

We recall that, via the **virial theorem**, the pressure in a Lennard-Jones system can be estimated by means of:

$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

which can be aptly converted to reduced units:
$$ P^\star = \rho^\star T^\star + \frac{1}{3V^\star} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\left[ \left(\frac{1}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{1}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle =
\rho^\star T^\star + \frac{\langle W \rangle}{3V^\star}$$
