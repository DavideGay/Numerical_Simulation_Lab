# Exercise 6

### The Ising model

The goal of this exercise is to study the properties of the 1D Ising model using the Gibbs and metropolis algorithms.

The Ising model describes a system of N spins (having values $\pm 1$) in a D dimensional lattice, each interacting with its closest neighbors.
The Hamiltonian for the 1D system is the following:
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$
where $J$ is the parameter controlling the intensity of the interaction.

The 1D Ising model is analitically solvable.
Its partition function can be computed and is
$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.

All relevant properties (internal energy, magnetization, magnetic susceptibility, specific heat) can be calculated exactly from the partition function.

### The sampling algorithms

To simulate the evolution of a Ising system, this exercise will use the Metropolis and Gibbs sampling algorithms.

We recall that the Ising system is a canonical ensemble, that is a system in thermal equilibrium with a heat bath at a fixed temperature.

- The **Metropolis algorithm** works as follows:

First, it picks a random spin and 'proposes' to modify its value.

The algorithm then calculates the difference of energy $\Delta E$ between the configuration that would result from accepting the move and the original configuration.

If $\Delta E<0$ the algorithm accepts the move, as the new configuration is energetically favored.

If $\Delta E>0$ the move is accepted with probability $exp(-\beta \Delta E)$, where $\beta=1/T$

- The **Gibbs algorithm** works as follow:

The algorithm randomly selects a spin from configuration and proposes a new value for it, independent of its value in the previous configuration.

The new spin value is drawn from the conditional distribution defined by the neighboring spins:
$$
p(s_k=\pm 1)=\frac{1}{1-exp(-\beta \Delta E)}
$$
where $\Delta E$ is the difference in energy resulting from updating the k-th spin.

Note that the Gibbs algorithm always accepts the proposed move.

## EQUILIBRATION

We will investigte the property of a system of $N=50$ spins.

First, we must ensure that the system reaches an **equilibrium** before computing any relevant physical quantities.
But how many simulation steps are required to reach an equilibrium configuration?
To tackle this problem, we will plot values of internal energy at different temperatures $(T=2,1.5,1,0.5)$ and using both Gibbs and Metropolis sampling.

For this simulations, we will use 5000 blocks of 1 step each.
The `equilibration.sh` script will take care of generating all the necessary data.
For each temperature, a new system of randomly distributed spins is generated.

```python
temps_array=[2,1.5,1,0.5]
n_part=50
```

```python
fig, axs = plt.subplots(2, 1, figsize = (10, 10))
cut=500
for t in range(len(temps_array)):
    x, en_inst, en_ave, en_err= np.loadtxt('../es6/metropolis/equilibration/total_energy_' + str(temps_array[t]) + '.dat',usecols=(0,1,2,3), unpack=True)
    axs[1].plot(x[:cut], en_inst[:cut], label='$ T = $' + str(temps_array[t]))
    axs[0].errorbar(x, en_ave,en_err, label='$ T = $' + str(temps_array[t]))
for i in range(2):
    axs[i].set_xlabel('Blocks')
    axs[i].set_ylabel('U/N')
    axs[i].legend(loc="upper right")
    axs[i].grid(True)
axs[1].set_title('Instant internal energy, first ' +str(cut)+' values')
axs[0].set_title('Moving average internal energy')
plt.show()
```

We can see that the system quickly reaches a somewhat steady average energy value, although it is subject to a great amount of fluctuations. We decide that 3000 simulation steps are surely enough to reach an equilibrium configuration.

We would also like to know whether restarting the system from a previous equilibrium configuration will affect the equilibration time.
To do so, the `restart.sh` script operates a series of N "coolings" starting from a temperature $T=2$ all the way down to $T=0.5$, at N regular intervals. For each temperature, 5000 simulation steps will be performed, then the final configuration will used as starting configuration for the next run at lower temperature.

We will then compare the internal energy of the system with that obtained from a random configuration, as function of the simulation steps.

```python
plt.figure(figsize=(10,10))
N=[5,10,20]
alg=['gibbs', 'metropolis']

for a in range(len(alg)):
    plt.subplot(2, 1, a+1)
    x, int_en, en_err= np.loadtxt('../es6/'+str(alg[a])+'/equilibration/total_energy_0.5.dat',usecols=(0,2, 3), unpack=True)
    plt.errorbar(x, int_en, en_err, label='$ T = $' + str(temps_array[t]))
    for i in range(len(N)):
        x, int_en, en_err= np.loadtxt('../es6/'+str(alg[a])+'/equilibration/total_energy_restart_'+str(N[i])+'.dat',usecols=(0,2, 3), unpack=True)
        plt.errorbar(x, int_en, en_err, label='$ T = $' + str(temps_array[t])+ ', '+str(N[i])+' coolings')

    plt.xlabel('Blocks')
    plt.ylabel('U/N')
    plt.legend(loc="upper right")
    plt.grid(True)
    plt.title('Internal energy (h=0), '+alg[a])

plt.show()
```

Judging from the graph above, starting from a previous configuration with a similar temperature does seem to provide negligible benefits, as an equilibrium appears to be reached after around the same amount of steps.

This result can be explained by the fact that, as we can see in the previous plots, the system's energy at each simulation step is extremely volatile due to fluctuations that only disappear as the the temperature approaches zero. 

Nonetheless, I decided to restart each new simulation from the final configuration of the previous simulation, since this only requires the addition of a couple of lines in the `measures.sh` script.

# RESULTS

We will measure relevant properties of the system as function of the temperature.
For each property, and for each temperature in the designed range, a simulation using 20 blocks of 20000 steps will be performed, at the end of which the final value obtained will be registered. All properties will be measured using a value $h=0$, except the magnetization with $h=0.02$.

### Internal energy

The internal energy per spin can be calculated analitically:

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
points=20
T = np.linspace(2.0,0.5,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

plt.figure(figsize = (12,8))

temp, tenergy_gibbs, tenergy_err_gibbs=np.loadtxt('../es6/gibbs/temp_tenergy.dat', usecols = (0, 1, 2), unpack = True)
tenergy_metro, tenergy_err_metro=np.loadtxt('../es6/metropolis/temp_tenergy.dat', usecols = (1, 2), unpack = True)

plt.subplot(2, 1, 1)
plt.plot(T, e, label='Analytic', color='r')
plt.errorbar(temp, tenergy_gibbs, yerr=tenergy_err_gibbs, linestyle='none' , marker='o', label='Gibbs')
plt.errorbar(temp, tenergy_metro, yerr=tenergy_err_metro, linestyle='none' , marker='o', label='Metropolis')
plt.title('Ising 1D, internal energy')
plt.legend()
plt.xlabel('$T$')
plt.ylabel('$U/N$')

plt.subplot(2, 1, 2)
plt.errorbar(temp, tenergy_gibbs - e, yerr = tenergy_err_gibbs, marker = 'o', label = "Gibbs")
plt.errorbar(temp, tenergy_metro - e, yerr = tenergy_err_metro, marker = 'o', label = "Metropolis")
plt.title("Errors")
plt.ylabel('$\Delta U/N$')
plt.xlabel('$T$')
plt.legend()
plt.axhline(y = 0, c = 'red')
plt.subplots_adjust(hspace = 0.5)
plt.show()
```

### Magnetization

The magnetization (with $h \not= 0$):

$$
M(N,T,h) =  \left\langle \sum_{i=1,N}s_i \right\rangle
=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

plt.figure(figsize = (12,8))

temp, magn_gibbs, magn_err_gibbs=np.loadtxt('../es6/gibbs/temp_magnetization.dat', usecols = (0, 1, 2), unpack = True)
magn_metro, magn_err_metro=np.loadtxt('../es6/metropolis/temp_magnetization.dat', usecols = (1, 2), unpack = True)

plt.subplot(2, 1, 1)
plt.plot(T, M, label='Analytic', color='r')
plt.errorbar(temp, magn_gibbs, yerr=magn_err_gibbs, linestyle='none' , marker='o', label='Gibbs')
plt.errorbar(temp, magn_metro, yerr=magn_err_metro, linestyle='none' , marker='o', label='Metropolis')
plt.title('Ising 1D, magnetization')
plt.legend()
plt.xlabel('$T$')
plt.ylabel('$U/N$')

plt.subplot(2, 1, 2)
plt.errorbar(temp, magn_gibbs - M, yerr = magn_err_gibbs, marker = 'o', label = "Gibbs")
plt.errorbar(temp, magn_metro - M, yerr = magn_err_metro, marker = 'o', label = "Metropolis")
plt.title("Errors")
plt.ylabel('$\Delta M/N$')
plt.xlabel('$T$')
plt.legend()
plt.axhline(y = 0, c = 'r')
plt.subplots_adjust(hspace = 0.5)
plt.show()
```

### Heat capacity

- the Heat Capacity:

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

plt.figure(figsize = (12,8))

temp, heatc_gibbs, heatc_err_gibbs=np.loadtxt('../es6/gibbs/temp_specific_heat.dat', usecols = (0, 1, 2), unpack = True)
heatc_metro, heatc_err_metro=np.loadtxt('../es6/metropolis/temp_specific_heat.dat', usecols = (1, 2), unpack = True)

plt.subplot(2, 1, 1)
plt.plot(T, heat, label='Analytic', color='r')
plt.errorbar(temp, heatc_gibbs, yerr=heatc_err_gibbs, linestyle='none' , marker='o', label='Gibbs')
plt.errorbar(temp, heatc_metro, yerr=heatc_err_metro, linestyle='none' , marker='o', label='Metropolis')
plt.title('Ising 1D, Heat capacity')
plt.legend()
plt.xlabel('$T$')
plt.ylabel('$C$')

plt.subplot(2, 1, 2)
plt.errorbar(temp, heatc_gibbs - heat, yerr = heatc_err_gibbs, marker = 'o', label = "Gibbs")
plt.errorbar(temp, heatc_metro - heat, yerr = heatc_err_metro, marker = 'o', label = "Metropolis")
plt.title("Errors")
plt.ylabel('$C$')
plt.xlabel('$T$')
plt.legend()
plt.axhline(y = 0, c = 'r')
plt.subplots_adjust(hspace = 0.5)
plt.show()
```

### Magnetic susceptibility

- The magnetic susceptibility:

$$
\chi(N,T) = \frac{\partial M}{\partial h} = \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right]  = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

plt.figure(figsize = (12,8))

temp=np.loadtxt('../es6/gibbs/temp_susceptibility.dat', usecols = 0)
# data is to be divided by n_part
chi_gibbs, chi_err_gibbs=np.loadtxt('../es6/gibbs/temp_susceptibility.dat', usecols = (1, 2), unpack = True)/n_part
chi_metro, chi_err_metro=np.loadtxt('../es6/metropolis/temp_susceptibility.dat', usecols = (1, 2), unpack = True)/n_part

plt.subplot(2, 1, 1)
plt.plot(T, X, label='Analytic', color='r')
plt.errorbar(temp, chi_gibbs, yerr=chi_err_gibbs, linestyle='none' , marker='o', label='Gibbs')
plt.errorbar(temp, chi_metro, yerr=chi_err_metro, linestyle='none' , marker='o', label='Metropolis')
plt.title('Ising 1D, Susceptibility')
plt.legend()
plt.xlabel('$\chi$')
plt.ylabel('$C$')

plt.subplot(2, 1, 2)
plt.errorbar(temp, chi_gibbs - X, yerr = chi_err_gibbs, marker = 'o', label = "Gibbs")
plt.errorbar(temp, chi_metro - X, yerr = chi_err_metro, marker = 'o', label = "Metropolis")
plt.title("Errors")
plt.ylabel('$\chi$')
plt.xlabel('$T$')
plt.legend()
plt.axhline(y = 0, c = 'r')
plt.subplots_adjust(hspace = 0.5)
plt.show()
```
