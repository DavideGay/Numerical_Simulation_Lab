# Exercise 9

The goal of this exercise is to tackle a few variations of the **Traveling Salesman Problem** (TSP) using a **Genetic Algorithm** (GA).

### The Traveling Salesman Problem

The **traveling salesman problem** (TSP) consists of a salesman and a set of $N$ cities with positions $\vec{c}_i$. The salesman has to visit each one of the cities and **must** start from and return to the same city, while **minimizing** the total length of the trip.

Consider now the ordered sequence of cities, $\vec{c}_1, \vec{c}_2, \dots, \vec{c}_N$, which goes from city 1 to the n$^{th}$ city. The path that minimizes the following **loss/cost function**:

$$
L(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|
$$


(with $\vec{x}_{N+1}=\vec{x}_1$) is one particular permutation, P, of the ordered sequence of cities $\vec{c}_1, \vec{c}_2, \dots, \vec{c}_N$, i.e. 

$$
\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N = \vec{c}_{P1}, \vec{c}_{P2}, \dots, \vec{c}_{PN}
$$

We can express a possible path via a 1D vector whose elements, from the first to the last, are integers that represents the cities in the order in which the salesman would visit them; for example for $N=5$ cities one possible path is represented by the vector:

$$
\left[ 1, 3, 5, 4, 2 \right]
$$

## Genetic Algorithm

To tackle the TSP we will use a Genetic Algorithm with features to be explained in this section.

Firstly, we create a `City` class, consisting in an object having as attributes the coordinates (x, y) of a specific city and, if needed (see Ex. 10), its name. To define a specific TSP we will have to initialize a vector of `City` objects describing the cities that the salesman needs to visit.

Then, we define a `Path` class, that will be used to identify a complete valid path connecting all the cities. It will have a vector of integers as attribute. These integers will each indicate one city, presented in the order they are visited. This class will have a number of methods, most notably one that returns the total distance covered by the path: each integer will be an index for the vector of Cities, thus giving access to each city's coordinates in the order given by the specific path. Other important methods will be functions that perform shifts, swaps, inversions (to be described soon) or shuffle.

Lastly, we define a `Population` class, having a vector of `Path` objects as attribute. This will basically be the "population" that the Genetic Algorithm acts on. For these ecercises, a population of 200 paths was chosen. Among the several methods given to this class are a **sorting operator**, some **mutation operators** and a **crossover operator**.

### GA: operators

- A `fitness_sort` operator, that allows the algorithm to sort the population based on the **loss function** i.e. the distance $L$ covered by each path, from most to least "fit".
- A `mutate` operator that, with a set probability, performs each mutation operator on each individual across the population:
    - a `mutation_pair` operator, that swaps two random cities in the path.
    - a `mutation_inversion` operator, that inverts the order in which m consecutive cities appear.
    - a `mutation_permutation` operator, that performs a random permutation of m cities.
    - a `mutation_shift` operator, that shifts forward m cities by n positions.
- A `select` operator, that, from an already sorted population, selects a "fit" individual for crossover with index $i=int(M*r^p)$, where M is the population size, r a random number (0,1] and p a suitable exponent >1. In this case, p was set as 5.
- A `crossover` operator, that selects two "parents" among the population and performs **crossover**:
    1. cuts their paths at the same position.
    2. conserves the first part of the paths.
    3. completes the paths with the missing cities adding them in the order in which they appear in the consort.

### GA: the algorithm

- First, the required set of cities is generated, and a population is initialized with completely random paths.
- The population is sorted based on the value of $L$ for each path.
- **"EVOLUTION"**: for each GA "generation", until the run is complete:
    1) Mutations are performed.
    2) Crossover is performed.
    3) The population is again sorted by fit.

### GA: parameters

Parameters needed to set up the GA code are provided from the user in the file `input.dat`.
Here is the list of all possible parameteres and the values that have been picked for this exercises.

- CITIES_TYPE:           0|1|2|3
- N_CITIES:              34
- POPULATION_SIZE:       200
- MUTATION PROBABILITY:  0.09
- N_GENERATIONS:         *(varies)*

The parameter CITIES_TYPE takes care of the generation of the coordinates of the cities for each scenario explored in these exercises.

The following functions will be helpful in calculating the best past for the simplest cases, allowing us to compare the GA results with the true best distance.

```python
def sort_points_by_angle(points):   # sorts array of positions based on their angle
    points_list = points.tolist()
    
    def angle_from_positive_x_axis(point):
        x, y = point
        angle= math.atan2(y, x)
        if angle <0:
            angle+=2*math.pi
        return angle
    
    sorted_points_list = sorted(points, key=angle_from_positive_x_axis)
    sorted_points = np.array(sorted_points_list)
    return sorted_points

def totdist(a,N):   # calculates total distance for array of positions. !!! Can eliminate N in favor of a[0,:].size
    totd=0
    def distance(a,b):
        return math.sqrt((b[0]-a[0])**2+(b[1]-a[1])**2)
    for i in range(N-1):
        totd+=distance(a[i,:],a[i+1,:])
    totd+=distance(a[N-1,:],a[0,:])
    return totd

def best_dist_circumference(a, N):   # sorts positions based on angle and returns total distance
    a_sorted = sort_points_by_angle(a)
    return totdist(a_sorted, N)
```

## Ex. 9.0 (test): Equally spaced cities over unit circumference

As a first attempt, we apply the GA to a TSP where all the cities are placed over the unit circumference, and are **equally spaced** (CITIES_TYPE 0).

What follows are scatter plots of the **best individual** in the first generation (before the GA run, first plot) and the last generation (after the GA run, second plot).

```python
pos = np.loadtxt('./DATA/test/cities.dat')

first_gen = []
last_gen = []
with open('./DATA/test/best_path_steps_0.dat', 'r') as file:
    lines = file.readlines()
    # Get the specific line, strip any leading/trailing whitespace, and split by space
    first_gen = lines[0].strip().split(" ")
    last_gen = lines[len(lines)-1].strip().split(" ")
    
first_path=np.zeros(shape=(34,2))
for i in range(len(first_gen)):
    city=int(first_gen[i])
    first_path[i,:]=pos[city,:]
    
last_path=np.zeros(shape=(34,2))
for i in range(len(last_gen)):
    city=int(last_gen[i])
    last_path[i,:]=pos[city,:]
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(first_path[:,0],first_path[:,1], marker='o')
plt.plot(first_path[0,0],first_path[0,1], marker='o')
plt.title(str(pos[:,0].size) + ' cities, first generation best')

plt.subplot(1,2,2)
plt.plot(last_path[:,0],last_path[:,1], marker='o')
plt.plot(first_path[0,0],first_path[0,1], marker='o')
plt.title(str(pos[:,0].size) + ' cities, last generation best')

plt.show()
```

We can clearly see that the path obtained is the best one.

Below we graph the distance as function of GA generations:
- in the first graph, the distance averaged over the "best" half of the population.
- in the second graph, the distance of the "best" path in the population.

We can see how less than 100 generations are required to reach the optimal value.

```python
best_half=np.loadtxt('./DATA/test/best_half_0.dat')
best_individual=np.loadtxt('./DATA/test/best_individual_0.dat')

x=np.arange(best_half.size)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].axhline(y=best_dist_circumference(pos, pos[:,0].size), color='r', linestyle='--', label='Optimal value')
axs[0].plot(x, best_half, label='Average distance over best half of population')
axs[0].set_xlabel('# Generation')
axs[0].set_ylabel('Total distance')
axs[0].legend(loc="upper right")
axs[0].grid(True)
axs[1].axhline(y=best_dist_circumference(pos, pos[:,0].size), color='r', linestyle='--', label='Optimal value')
axs[1].plot(x, best_individual, label='Best individual across population')
axs[1].set_xlabel('# Generation')
axs[1].set_ylabel('Total distance')
axs[1].legend(loc="upper right")
axs[1].grid(True)

print(best_dist_circumference(pos, pos[:,0].size))
```
