# Exercise 10

In this exercise, we will take advantage of *parallel computing* to (possibly) improve the performance of the **Genetic Algorithm** (GA) described in the previous exercise.

**Parallelization** consists in dividing a problem into subproblems that can be executed simultaneously on multiple processors or computing cores, allowing maximum exploitation of the available computational resources. 

For this exercise, we will use **MPI libraries**.

## Ex. 10.1

We want to tackle the TSP problem presented in the last section of Ex. 9 with the use of parallelization.
In particular, we will take advantage of running parallel GA algorithms (The "population" of each being called a **"continent"**) by implementing what we will call a **"migration"**:
- Every $N$ (in this case set as 20) GA generations, all the continents exchange their best individual with each other.

One could also proceed with a slightly different "migration" implementation, for instance by exchanging more than one individual and/or picking the "migrator(s)" using the `select` operator.

We will almost entirely use the GA code described in Ex. 9, which was already built to be easily adaptable to parallel processing. Relevant changes are limited to the the "main", with the addition of the required MPI setup and the "migration" code. A new input will be:
- MIGRATIONS 0|1

which will indicate whether we want to apply migrations (1) or simply run parallel GA runs (0).

The only issue that arises is the fact that, as the generation of cities coordinates is done independently for each process, when working with CITIES_TYPE 2 (random cities inside a square) each process will work on a different set of cities. To circumvent this issue, we will use CITIES_TYPE 3 (read coordinates from input file) using a cities configuration previously created by a CITIES_TYPE 2 run.

We now plot the progression of the total distance of the best path in each of 4 parallel processes "continents", comparing the results achieved without and with migrations. Each continent is made of 300 paths.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

for i in range(4):
    plt.subplot(2,2,i+1)
    best_nc=np.loadtxt('DATA/10_1/no_migration/best_individual_'+str(i)+'.dat')
    x=np.arange(best_nc.size)
    best_c=np.loadtxt('DATA/10_1/with_migration/best_individual_'+str(i)+'.dat')
    plt.plot(x, best_nc, label='Non communicating processes, best individual')
    plt.plot(x, best_c, label='Communicating processes, best individual')
    plt.xlabel('# Generation')
    plt.ylabel('Total distance')
    plt.legend()
    plt.grid(True)
plt.show()
```

As we can see, no relevant improvement of the results has been achieved with regard to this TSP. On the contrary, the light "shuffling" caused by the exchanging of the best individual seems to slightly slow down the convergence of the algorithm.

Nonetheless, we would like to test it with a more difficult TSP - one with many more cities to connect. We expect to be able to see slight but noticeable improvements.

## Ex. 10.2

In this section we will apply the GA to solve a TSP in which the cities are provided by a file containing a list of 110 Italian cities along with their "real world" coordinates.

```python
pos = np.loadtxt('DATA/10_2/no_migration/cities.dat')
```

```python
plt.figure(figsize=(5,5))
plt.scatter(pos[:,0], pos[:,1], label='Cities')
plt.legend()
plt.grid(True)
plt.title('110 Italian cities')
plt.show()
```

We will proceed in the same way as for Ex. 10.1.

## Non communicating parrallel processes

First we want to visualize results achieved without communication between the 4 processes, with continents of 500 individuals (i.e. paths) each.

Plots below show the best path obtained at the end of each parallel GA run.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

for i in range(4):
    last_gen = []
    with open('DATA/10_2/no_migration/best_path_steps_'+str(i)+'.dat', 'r') as file:
        lines = file.readlines()
        # Get the specific line, strip any leading/trailing whitespace, and split by space
        last_gen = lines[len(lines)-1].strip().split(" ")

    last_path=np.zeros(shape=(len(last_gen),2))
    for j in range(len(last_gen)):
        city=int(last_gen[j])
        last_path[j,:]=pos[city,:]
    
    plt.subplot(2,2,i+1)
    plt.plot(last_path[:,0], last_path[:,1], marker='o', label="Process #" + str(i) + ", best path")
    plt.plot(last_path[0,0], last_path[0,1], marker='o')
    plt.grid(True)
    plt.legend()
    
    best=np.loadtxt('DATA/10_2/no_migration/best_individual_'+str(i)+'.dat')
    print("Process #" + str(i) + ", best distance: " + str(best[-1]))
plt.show()
```

Here we graphed the convergence of the best individual's distance:

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i in range(4):
    best=np.loadtxt('DATA/10_2/no_migration/best_individual_'+str(i)+'.dat')
    x=np.arange(best.size)
    
    plt.subplot(2,2,i+1)
    plt.plot(x, best, label='#' +str(i)+', non comm. processes, best individual')
    plt.xlabel('# Generation')
    plt.ylabel('Total distance')
    plt.grid(True)
    plt.legend()
    print("Process #" + str(i) + ", best distance: " + str(best[-1]))
    
plt.show()
```

## Communicating processes

Now we show the results obtained with the implementation of "migrations".

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

for i in range(4):
    last_gen = []
    with open('DATA/10_2/with_migration/best_path_steps_'+str(i)+'.dat', 'r') as file:
        lines = file.readlines()
        # Get the specific line, strip any leading/trailing whitespace, and split by space
        last_gen = lines[len(lines)-1].strip().split(" ")

    last_path=np.zeros(shape=(len(last_gen),2))
    for j in range(len(last_gen)):
        city=int(last_gen[j])
        last_path[j,:]=pos[city,:]
    
    plt.subplot(2,2,i+1)
    plt.plot(last_path[:,0], last_path[:,1], marker='o', label="Process #" + str(i) + ", best path")
    plt.plot(last_path[0,0], last_path[0,1], marker='o')
    plt.grid(True)
    plt.legend()
    
    best=np.loadtxt('DATA/10_2/with_migration/best_individual_'+str(i)+'.dat')
    print("Process #" + str(i) + ", best distance: " + str(best[-1]))
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i in range(4):
    best_nocomm=np.loadtxt('DATA/10_2/no_migration/best_individual_'+str(i)+'.dat')
    x=np.arange(best.size)
    best_comm=np.loadtxt('DATA/10_2/with_migration/best_individual_'+str(i)+'.dat')
    
    plt.subplot(2,2,i+1)
    plt.plot(x, best_nocomm, label='#' +str(i)+', non comm. processes, best individual')
    plt.plot(x, best_comm, label='#' +str(i)+', comm. processes, best individual')
    plt.xlabel('# Generation')
    plt.ylabel('Total distance')
    plt.grid(True)
    plt.legend()
    
plt.show()
```

We can see that, in this case, taking advantage of parallel computing has lead to slightly better results.

Again, the convergence is slightly slower in the first parts of the run, but after 100-200 generations the communicating runs produce the better results.
