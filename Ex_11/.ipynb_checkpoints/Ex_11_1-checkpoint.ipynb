```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear fit

We consider the simple case of data $(x,y)$ obtained from 
$$
    y_i= f(x_i) + \sigma_i,
$$
where
$$
f(x)=2x+1
$$
and $\eta$ is uncorrelated Gaussian noise.

First we define the linear function parameters:

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function
```

```python
sigma = 0.2 # noise standard deviation
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

To do a linear fit, only one neuron is required: weigth*inputdata + bias

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))   # add 1 layer composed of 1 neuron

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

We are now going to train our model, that is we feed the neuron with the set of training pair x, y_train from which the optimizer will find the best weights to minimize the Mean Square Error loss function.

```python
# fit the model using training dataset
# over 30 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

Weight is m parameter; bias is b:

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

From the results obtained, we can see that after a certain number of Epochs (around 15) the loss is stable and little to no improvement is made by training the model further.

We now wish to investigate how the values of $N_{train}$ and $\sigma$ impact the results.

```python
noise=[0., 0.1, 0.2, 0.3]
Ntrain=[100, 1000, 5000]
fig, axs = plt.subplots(4, 3, sharey=False, tight_layout=True, figsize=(12, 16))

for j in range(4):
    for i in range(3):
        x_train = np.random.uniform(-1, 1, Ntrain[i])
        x_valid = np.random.uniform(-1, 1, 50)
        x_valid.sort()
        y_target = m * x_valid + b # ideal (target) linear function

        y_train = np.random.normal(m * x_train + b, noise[j]) # actual measures from which we want to guess regression parameters
        y_valid = np.random.normal(m * x_valid + b, noise[j])
        
        # create the model
        model = tf.keras.Sequential()
        model.add(Dense(1, input_shape=(1,)))   # add 1 layer composed of 1 neuron

        # compile the model choosing optimizer, loss and metrics objects
        model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
        
        history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=15, verbose=0,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
        
        axs[j,i].plot(history.history['loss'])
        axs[j,i].plot(history.history['val_loss'])
        axs[j,i].set_ylabel('Loss')
        axs[j,i].set_xlabel('Epoch')
        axs[j,i].legend(['Train', 'Test'], loc='best')
        axs[j,i].set_title("$\sigma$ = "+str(noise[j])+", $N_{train}$ = "+str(Ntrain[i]))
        axs[j,i].grid(True)

plt.show()
```

From these results we can observe that, in order to learn from data with more noise, a larger training data set is required.
Also, the number of epochs required to reach an adequate value of loss decreases significantly with the increase of training data points.
