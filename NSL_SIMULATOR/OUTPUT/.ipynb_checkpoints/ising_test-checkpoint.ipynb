```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e)
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.show()
```

```python
filepath='temp_tenergy.dat'

temp=np.loadtxt(filepath, usecols = (0, ))
ave=np.loadtxt(filepath, usecols = (1, ))
err=np.loadtxt(filepath, usecols = (2, ))
print(err)
```

```python
points=100
T = np.linspace(0.2,3.,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e)
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.errorbar(temp, ave, yerr=err, fmt="o")
plt.grid(True)
plt.show()
```
